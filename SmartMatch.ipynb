```python
import numpy as np
from faker import Faker
import random

fake = Faker()
num_freelancers = 1000
names = [fake.name() for _ in range(num_freelancers)]


freelancer_matrix = np.array([[round(random.uniform(1, 10), 1) for _ in range(3)] for _ in range(num_freelancers)])

print("Matrix Shape:", freelancer_matrix.shape)
print("First Freelancer:", names[0], "-> Vector:", freelancer_matrix[0])
print(freelancer_matrix)
```

# Define Cosin Similarity

```python
target = np.array([9.0, 4.0, 8.0])

dot_product = freelancer_matrix @ target

matrix_norms = np.linalg.norm(freelancer_matrix, axis=1)

target_norm = np.linalg.norm(target)

scores = dot_product / (matrix_norms * target_norm)
print(scores[868])

print(scores[362])


top_3_indices = np.argsort(scores)[::-1][:3]
print(top_3_indices)

print("--- Top 3 Best Matches (Cosine) ---")
for idx in top_3_indices:
    print(f"Name: {names[idx]} | Vector: {freelancer_matrix[idx]} | Match Score: {scores[idx]:.4f}")


```
